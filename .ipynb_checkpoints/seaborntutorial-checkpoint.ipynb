```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sns.set_style("whitegrid")
sinplot()
```

```python
sns.set(rc={"figure.figsize": (6,6)})
np.random.seed(sum(map(ord, "palettes")))
```

```python
# qualitative or categorical palettes are best for categories
# that do not have any inherent ordering

# default color palette is a set of six colors
current_palette = sns.color_palette()
# palplot stands for pallete plot
sns.palplot(current_palette)
```

```python
# see all types of available palettes
paltypes = ["deep", "muted", "bright", "pastel", "dark", "colorblind"]
for t in paltypes:
    palette = sns.color_palette(t)
    sns.palplot(palette)
```

```python
sns.palplot(sns.color_palette("husl", 8))
```

```python
sns.palplot(sns.color_palette("husl", 10))
```

```python
sns.palplot(sns.color_palette("husl", 20))
```

```python
# tips is an existing dataset 
# within seaborn loaded with load_dataset
tips = sns.load_dataset("tips")
```

```python
# shows histogram (normed meaning total area is 1)and kde
# useful to see distribution of single array, series etc
sns.distplot(tips.total_bill)
```

```python
# by default hist=True and kde=True, 
# now we also turn on rug=True

sns.distplot(tips.total_bill, rug=True, hist=True, kde=True)
```

```python
# not show the kde and not-norm the histogram

sns.distplot(tips.total_bill, rug=True, hist=True, kde=False, norm_hist=False)
```

```python
# kde plots pdf
ax = sns.kdeplot(tips.total_bill)
```

```python
# kde plot with shading
ax = sns.kdeplot(tips.total_bill, shade=True)
```

```python
# two dimensional gaussian seen, uses sequential maps for contours
ax = sns.kdeplot(tips.total_bill, tips.tip)
```

```python
# number of levels for 2d gaussian
ax = sns.kdeplot(tips.total_bill, tips.tip, n_levels=20)
```

```python
# kde has something known as kernel and bandwidth
ax = sns.kdeplot(tips.total_bill, kernel="cos", bw="silverman")
```

```python
# show cdf instead of pdf
ax = sns.kdeplot(tips.total_bill, cumulative=True)
```

```python
# adding multiple plots one after other

sns.kdeplot(tips.total_bill)
sns.rugplot(tips.total_bill)
```

```python
# now we see pair plots
iris = sns.load_dataset("iris")
```

```python
# scatter plot matrix, useful for looking at correlations
# scatter plots are also known as joint distributions
print iris.head()
g =sns.pairplot(iris)
```

```python
# different categories (of species) in histogram are colored
g = sns.pairplot(iris, hue="species")
```

```python
# change markers in the above plot
g = sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
```

```python
# show scatterplot/pairplot with regression
g = sns.pairplot(iris, hue="species", kind="reg")
```
